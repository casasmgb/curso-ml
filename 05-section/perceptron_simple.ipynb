```python
import numpy as np
import matplotlib.pyplot as plt

# Crear un conjunto de datos de clasificacin simple
np.random.seed(42)
X = np.array([[2, 2], [3, 4], [4, 3], [6, 5], [8, 7], [10, 10]])
y = np.array([0, 0, 0, 1, 1, 1])

# Funcin de activacin ReLU
def relu(x):
    return max(0, x)

# Funcin de activacin (umbral)
def step_function(x):
    return 1 if x >= 0 else 0

# Inicializacin de pesos y sesgo
weights = np.array([0.5, -0.5])
bias = 0

# Hiperparmetros
learning_rate = 0.1
epochs = 10

# Entrenamiento del perceptrn con funcin de activacin ReLU
losses = []
accuracies = []

for epoch in range(epochs):
    total_loss = 0
    correct_predictions = 0
    for i in range(len(X)):
        z = np.dot(weights, X[i]) + bias  # Calcula la suma ponderada (z)
        prediction = relu(z)  # Aplica la funcin de activacin ReLU
        # prediction = step_function(np.dot(weights, X[i]) + bias) # Aplica la funcin de activacin step
        error = y[i] - prediction  # Calcula el error
        weights += learning_rate * error * X[i]  # Ajusta los pesos basado en el error y los datos
        bias += learning_rate * error  # Ajusta el sesgo basado en el error
        total_loss += error ** 2
        if prediction == y[i]:
            correct_predictions += 1
    losses.append(total_loss)
    accuracy = correct_predictions / len(X)
    accuracies.append(accuracy)

# Visualizacin de la funcin de prdida y la precisin
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(1, epochs + 1), losses, marker='o')
plt.xlabel('poca')
plt.ylabel('Prdida')
plt.title('Funcin de Prdida durante el Entrenamiento con ReLU')

plt.subplot(1, 2, 2)
plt.plot(range(1, epochs + 1), accuracies, marker='o', color='r')
plt.xlabel('poca')
plt.ylabel('Precisin')
plt.title('Precisin durante el Entrenamiento con ReLU')

plt.tight_layout()
plt.show()
```
