```python
import numpy as np
import matplotlib.pyplot as plt

# Crear un conjunto de datos de clasificación simple
np.random.seed(42)
X = np.array([[2, 2], [3, 4], [4, 3], [6, 5], [8, 7], [10, 10]])
y = np.array([0, 0, 0, 1, 1, 1])

# Función de activación ReLU
def relu(x):
    return max(0, x)

# Función de activación (umbral)
def step_function(x):
    return 1 if x >= 0 else 0

# Inicialización de pesos y sesgo
weights = np.array([0.5, -0.5])
bias = 0

# Hiperparámetros
learning_rate = 0.1
epochs = 10

# Entrenamiento del perceptrón con función de activación ReLU
losses = []
accuracies = []

for epoch in range(epochs):
    total_loss = 0
    correct_predictions = 0
    for i in range(len(X)):
        z = np.dot(weights, X[i]) + bias  # Calcula la suma ponderada (z)
        prediction = relu(z)  # Aplica la función de activación ReLU
        # prediction = step_function(np.dot(weights, X[i]) + bias) # Aplica la función de activación step
        error = y[i] - prediction  # Calcula el error
        weights += learning_rate * error * X[i]  # Ajusta los pesos basado en el error y los datos
        bias += learning_rate * error  # Ajusta el sesgo basado en el error
        total_loss += error ** 2
        if prediction == y[i]:
            correct_predictions += 1
    losses.append(total_loss)
    accuracy = correct_predictions / len(X)
    accuracies.append(accuracy)

# Visualización de la función de pérdida y la precisión
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(1, epochs + 1), losses, marker='o')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.title('Función de Pérdida durante el Entrenamiento con ReLU')

plt.subplot(1, 2, 2)
plt.plot(range(1, epochs + 1), accuracies, marker='o', color='r')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.title('Precisión durante el Entrenamiento con ReLU')

plt.tight_layout()
plt.show()
```
